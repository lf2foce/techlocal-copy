```python
from dotenv import load_dotenv
from google import genai
from google.genai import types
from pydantic import BaseModel
import asyncio
from typing import Dict, Any, List
import os, json

load_dotenv()

class Theme(BaseModel):
    title: str
    story: str

class ThemeGenerate(BaseModel):
    themes: List[Theme]
    mindset: str   
    plan: str

def generate_theme_title_and_story(campaign_title: str, insight: str, description: str, target_customer:str) -> List[tuple[str, str]]:
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
    
    # Generate response using Gemini API (synchronous version)
    response = client.models.generate_content(
        model='gemini-2.0-flash',
        contents=f"Lập kế hoạch triển khai hệ thống kênh social mới để bùng nổ cùng chiến dịch nam tính mới toàn cầu của oniiz.", # Tạo 5 thương hiệu cho pages với các thông tin {insight} {target_customer}. Mỗi thương hiệu phải có title và story khác nhau.
        config=types.GenerateContentConfig(
            response_mime_type='application/json',
            response_schema=ThemeGenerate,
            system_instruction=types.Part.from_text(text=system_prompt.format(desc=description, target_customer=target_customer, insight=insight)),
        ),
    )
    
    # Extract the response
    print("Generated 5 themes based on user prompt.")
    content = json.loads(response.text)
    
    # Validate and parse the response using Pydantic
    themes_data = ThemeGenerate(**content)
    
    # Convert list of themes to list of tuples using dot notation for Pydantic model attributes
    return themes_data.mindset, themes_data.plan
    return [(theme.title, theme.story) for theme in themes_data.themes]
```

```python
from typing import List, Literal, Optional
from pydantic import BaseModel
from datetime import date, timedelta
import random
import os
import json
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()

# --- STRUCTS ---
class MindsetSection(BaseModel):
    title: str
    content: str
#1
class PageSetup(BaseModel):
    page_name: str
    intro_story: str

class ContentItem(BaseModel):
    day: int
    content_type: Literal["quote", "reel", "confession", "story", "photo"]
    title: str
    description_ref: str

class GeneratedContent(BaseModel):
    day: int
    content_type: str
    caption: str
    hashtags: Optional[str]
    idea: Optional[str]
    image_prompts: Optional[List[str]] = None
#1
class BrandTheme(BaseModel):
    name: str
    manifesto: str

#1
class StrategyInputResponse(BaseModel):
    themes: List[BrandTheme]
    mindset: str
    plan: str

#1
class StrategyBundle(BaseModel):
    insight: str
    target_customer: str
    tone_of_voice: str
    page_setup: PageSetup
    mindset: List[MindsetSection]
    content_plan: List[ContentItem]

# --- FUNCTION 1: Generate Mindset, Page, Plan using Gemini ---
def generate_strategy_bundle_gemini(campaign_title: str, insight: str, description: str, target_customer: str) -> StrategyBundle:
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

    system_prompt = f"""
    Bạn là chiến lược gia nội dung chuyên về xây dựng chiến dịch xã hội.

    Hãy tạo kế hoạch xây dựng hệ thống kênh social cho chiến dịch \"{campaign_title}\".
    Chiến dịch này hướng đến đối tượng: {target_customer}.
    Insight cốt lõi: {insight}.

    Hãy dựa trên phong cách làm nội dung giống như sau:
    - Không còn thời của spam tool, nội dung cần có chiến lược, insight đúng, cảm xúc thật.
    - Xây page có định vị rõ ràng, kết nối với nhu cầu cảm xúc.
    - Nội dung reels, ảnh, quote đều gợi cảm xúc sâu sắc.
    - Content có thể dùng để tăng tương tác, kéo traffic hoặc gắn link POD.

    Trả kết quả JSON gồm 3 phần:
    - mindset: tư duy cảm xúc của chiến dịch
    - plan: kế hoạch triển khai nội dung
    - themes: danh sách 5 page với name và manifesto (emotional brand statement)
    """

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=f"Chiến dịch: {campaign_title}",
        config=types.GenerateContentConfig(
            response_mime_type="application/json",
            response_schema=StrategyInputResponse,
            system_instruction=types.Part.from_text(text=system_prompt)
        ),
    )

    data = StrategyInputResponse(**json.loads(response.text))

    mindset_section = [MindsetSection(title="Tư duy chiến dịch", content=data.mindset)]
    themes = data.themes
    page_setup = PageSetup(page_name=themes[0].name, intro_story=themes[0].manifesto)

    #    "quote": "Trích dẫn truyền cảm hứng giúp nam giới nhìn nhận cảm xúc một cách tích cực.",
    #     "reel": "Reel ngắn gợi cảm xúc hoặc phản ánh một khoảnh khắc chân thật của đàn ông.",
    #     "confession": "Chia sẻ thật, ẩn danh về áp lực mà nam giới thường không dám nói ra.",
            # "photo": "Ảnh giàu cảm xúc với lời nhắn chạm tới tâm lý người xem."

    types_list = [ "story"] # "quote", "reel", "confession", , "photo"
    preset_descriptions = {
     
        "story": "Một câu chuyện đời thực hoặc hư cấu khơi gợi sự đồng cảm, độ dài ít nhất 300 từ.",
    }

    random.shuffle(types_list)
    repeated_types = (types_list * 6)[:30]

    content_plan = [
        ContentItem(
            day=i+1,
            content_type=ctype,
            title=f"Ngày {i+1} - Nội dung {ctype.title()}",
            description_ref=preset_descriptions[ctype]
        )
        for i, ctype in enumerate(repeated_types)
    ]

    return StrategyBundle(
        insight=insight,
        target_customer=target_customer,
        tone_of_voice="Đồng cảm, truyền cảm hứng, hiện đại nhưng không sướt mướt",
        page_setup=page_setup,
        mindset=mindset_section,
        content_plan=content_plan
    )

# --- FUNCTION 2: Generate Daily Content using Gemini ---
def generate_ai_content_gemini(item: ContentItem, strategy: StrategyBundle) -> GeneratedContent:
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

    mindset_text = "\n".join([f"{s.title}: {s.content}" for s in strategy.mindset])

    system_prompt = f"""
    Bạn là chuyên gia sáng tạo nội dung cho chiến dịch \"{strategy.page_setup.page_name}\" – một chiến dịch toàn cầu hướng đến nam giới hiện đại.

    Chiến dịch tập trung vào insight: \"{strategy.insight}\", dành cho {strategy.target_customer}.
    Tone-of-voice: {strategy.tone_of_voice}.

    Tuyên ngôn thương hiệu:
    {strategy.page_setup.intro_story}

    Tư duy chiến dịch:
    {mindset_text}

    Ghi nhớ:
    - caption có thể ngắn, mang tính hook.
    - nếu content_type là \"story\" thì viết bài dài ít nhất 300 từ như một bài blog.
    - nếu content_type là \"reel\" hoặc liên quan video thì tạo thêm một trường idea mô tả chi tiết để dùng sinh hình ảnh AI (image_prompt).

    Trả kết quả JSON:
    {{
      "caption": "...",
      "hashtags": "...",
      "idea": "..."
    }}
    """

    user_prompt = f"""
    - Ngày: {item.day}
    - Dạng: {item.content_type}
    - Chủ đề: {item.title}
    - Nội dung theo định hướng: {item.description_ref}
    """

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=user_prompt,
        config=types.GenerateContentConfig(
            response_mime_type="application/json",
            response_schema=GeneratedContent,
            system_instruction=types.Part.from_text(text=system_prompt)
        ),
    )

    raw = json.loads(response.text)

    if item.content_type in ["story", "reel"] and raw.get("idea"):
        base_prompt = raw["idea"]
        raw["image_prompts"] = [
            f"{base_prompt}, góc nhìn toàn cảnh",
            f"{base_prompt}, góc nhìn cận cảnh",
            f"{base_prompt}, biểu cảm gương mặt",
            f"{base_prompt}, hành động đang diễn ra"
        ]

    return GeneratedContent(**raw)

# --- DEMO: In 3 ngày đầu tiên và gợi ý image_prompt nếu cần ---
# if __name__ == "__main__":
#     strategy = generate_strategy_bundle_gemini(
#         campaign_title="Nam Tính Mới",
#         insight="Đàn ông không nên chịu đựng một mình",
#         description="Chiến dịch nâng cao nhận thức về cảm xúc và sự kết nối của nam giới hiện đại",
#         target_customer="Nam giới 25–45 tuổi ở thành thị"
#     )

#     for i in range(3):
#         content = generate_ai_content_gemini(strategy.content_plan[i], strategy)
#         print(f"\n📅 Ngày {content.day} - [{content.content_type.upper()}]")
#         print("📌 Caption:", content.caption)
#         print("🏷️ Hashtags:", content.hashtags)
#         print("🧠 Idea:", content.idea)

#         if content.image_prompts:
#             for j, prompt in enumerate(content.image_prompts, 1):
#                 print(f"🖼️ Image Prompt {j}: {prompt}")
```

```python
#worked v1
from typing import List, Literal, Optional
from pydantic import BaseModel
from datetime import date, timedelta
import random
import os
import json
from dotenv import load_dotenv
from google import genai
from google.genai import types

load_dotenv()

# --- STRUCTS ---
class MindsetSection(BaseModel):
    title: str
    content: str

class PageSetup(BaseModel):
    page_name: str
    intro_story: str

class ContentItem(BaseModel):
    day: int
    content_type: Literal["quote", "reel", "confession", "story", "photo"]
    title: str
    description: str

class GeneratedContent(BaseModel):
    day: int
    content_type: str
    caption: str
    hashtags: Optional[str]
    idea: Optional[str]

class Theme(BaseModel):
    title: str
    story: str

class ThemeGenerate(BaseModel):
    themes: List[Theme]
    mindset: str
    plan: str

# --- FUNCTION 1: Generate Mindset, Page, Plan using Gemini ---
def generate_strategy_bundle_gemini(campaign_title: str, insight: str, description: str, target_customer: str):
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

    system_prompt = """
    Bạn là chiến lược gia nội dung chuyên về xây dựng chiến dịch xã hội.
    Hãy tạo kế hoạch cho một chiến dịch truyền thông mang tên \"{desc}\", tập trung vào insight \"{insight}\" và khách hàng mục tiêu \"{target_customer}\".

    Xuất kết quả dạng JSON với 3 phần:
    - mindset: tư duy định hướng chiến dịch
    - plan: kế hoạch triển khai
    - themes: danh sách 5 page có title + story
    """

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=f"Chiến dịch: {campaign_title}",
        config=types.GenerateContentConfig(
            response_mime_type="application/json",
            response_schema=ThemeGenerate,
            system_instruction=types.Part.from_text(text=system_prompt.format(
                desc=description,
                target_customer=target_customer,
                insight=insight,
            )),
        ),
    )

    print("Generated strategy bundle with mindset, plan, and themes.")
    content = json.loads(response.text)
    data = ThemeGenerate(**content)

    mindset_section = [MindsetSection(title="Tư duy chiến dịch", content=data.mindset)]
    plan_section = data.plan
    themes = data.themes
    page_setup = PageSetup(page_name=themes[0].title, intro_story=themes[0].story)

    types_list = ["quote", "reel", "confession", "story", "photo"]
    content_plan = []
    for i in range(1, 31):
        content_type = random.choice(types_list)
        title = f"Ngày {i} - Nội dung {content_type.title()}"
        description = f"{plan_section} - Dạng nội dung: {content_type}"
        content_plan.append(ContentItem(day=i, content_type=content_type, title=title, description=description))

    return {
        "mindset": mindset_section,
        "page_setup": page_setup,
        "content_plan": content_plan
    }

# --- FUNCTION 2: Generate Daily Content using Gemini ---
def generate_ai_content_gemini(item: ContentItem, mindset: List[MindsetSection]) -> GeneratedContent:
    client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

    mindset_text = "\n".join([f"{s.title}: {s.content}" for s in mindset])

    user_prompt = f"""
    Bạn là chuyên gia sáng tạo nội dung cho chiến dịch Nam Tính Mới. Dưới đây là tư duy chiến dịch:
    {mindset_text}

    Hãy tạo nội dung social media cho:
    - Ngày: {item.day}
    - Dạng: {item.content_type}
    - Chủ đề: {item.title}

    Xuất kết quả gồm:
    1. Caption cảm xúc
    2. Hashtags
    3. Gợi ý idea nếu là ảnh hoặc video
    """

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=user_prompt,
        config=types.GenerateContentConfig(response_mime_type="text/plain"),
    )

    return GeneratedContent(
        day=item.day,
        content_type=item.content_type,
        caption=response.text,
        hashtags="#NamTínhMới #KhôngMộtMình",
        idea=f"Idea for: {item.title}"
    )

# Example usage:
if __name__ == "__main__":
    STRATEGY = generate_strategy_bundle_gemini(
        campaign_title="Nam Tính Mới",
        insight="Đàn ông không nên chịu đựng một mình",
        description="Chiến dịch nâng cao nhận thức về cảm xúc và sự kết nối của nam giới hiện đại",
        target_customer="Nam giới 25–45 tuổi ở thành thị"
    )

    MINDSET = STRATEGY["mindset"]
    PLAN = STRATEGY["content_plan"]
    PAGE = STRATEGY["page_setup"]

    print("\n--- PAGE SETUP ---")
    print(PAGE)

    print("\n--- MINDSET ---")
    for s in MINDSET:
        print(f"{s.title}: {s.content}")

    print("\n--- DAY 1 GENERATED CONTENT ---")
    print(generate_ai_content_gemini(PLAN[0], MINDSET))
```
